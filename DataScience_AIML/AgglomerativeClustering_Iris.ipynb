```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
sb.set_style('dark')
df=pd.read_csv('IRIS.csv')
df.head(10)
```

```python
X2=df.iloc[:,1:10]
print(X2)
arr=np.array(X2)
print(arr)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(arr[:,0],arr[:,1],c='r')
plt.xlabel('X - COORDINATE')
plt.ylabel('Y - COORDINATE')
plt.title('SCATTER PLOT')
plt.xlim([0,10]),plt.ylim([0,10])
plt.xticks(range(10)),plt.yticks(range(10))
plt.grid()
plt.show()
```

```python
#Dendrogram
labellist=list(df['species'])
plt.figure(figsize=(15,6))
Z1=linkage(arr[:,0:2],method='ward')
dendrogram(Z1,labels=labellist,distance_sort='descending',show_leaf_counts=False)
plt.xlabel('Flowers')
plt.ylabel('Euclidean Distances')
plt.title('Dendrogram')
plt.show()
```

```python
#AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
cluster.fit(arr[:, 0:2])
# Scatter plot of clusters
fig, ax = plt.subplots(figsize=(6,6))
scatter = ax.scatter(arr[:,0], arr[:,1], c=cluster.labels_, cmap='rainbow')
legend = ax.legend(*scatter.legend_elements(), title='Clusters', bbox_to_anchor=(1,1))
ax.add_artist(legend)
plt.title('Scatter plot of clusters')
plt.show()
labels = cluster.labels_
print(labels)
```
