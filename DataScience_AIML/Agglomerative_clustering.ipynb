```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import linkage,dendrogram
#initial data points scatter plot
sns.set_style('dark')
X1=np.array([[1,1],[3,2],[9,1],[3,7],[7,2],[9,7],[4,8],[8,3],[1,4]])
plt.figure(figsize=(6,6))
plt.scatter(X1[:,0],X1[:,1],c='r')
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.title('Scatter plot of data')
plt.xlim([0,10]), plt.ylim([0,10])
plt.xticks(range(10)),plt.yticks(range(10))
plt.grid()
plt.show()
```

```python
#linkages
Z1=linkage(X1, method='single', metric='euclidean')
Z2=linkage(X1, method='complete', metric='euclidean')
Z3=linkage(X1, method='average',metric='euclidean')
print(Z1)
print(Z2)
print(Z3)
```

```python
#dendrogram
plt.figure(figsize=(15,10))
plt.subplot(2,2,1), dendrogram(Z1),plt.title('Single Linkage')
plt.subplot(2,2,2), dendrogram(Z2),plt.title('Complete Linkage')
plt.subplot(2,2,3), dendrogram(Z3),plt.title('Average Linkage')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
f1=fcluster(Z3,2,criterion='maxclust')
print(f"Clusters {f1}")
```

```python
#Agglomerative clustering
from sklearn.cluster import AgglomerativeClustering
Z1=AgglomerativeClustering(n_clusters=2,linkage='complete')
Z1.fit_predict(X1)
print(Z1.labels_)
```

```python
fig, ax = plt.subplots(figsize=(6,6))
scatter = ax.scatter(X1[:,0],X1[:,1],c=Z1.labels_,cmap='rainbow')
legend = ax.legend(*scatter.legend_elements(),title='Clusters',bbox_to_anchor=(1,1))
ax.add_artist(legend)
plt.title('Scatter plot of clusters')
plt.show()
```
