```python
#importing basic important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
#reading csv file
iris=pd.read_csv("IRIS.csv")
#printing elements from the head of the dataset within csv file
iris.head()
print(iris.head())
#printing 8 elements from the tail of the dataset within csv file
print(iris.tail(8))
#printing columns from IRIS.csv
iris.columns
print(iris.columns)
#Checking the dimensionality of the dataset
print(iris.shape)
#No. of points of each class
print(iris['species'].value_counts())
#basic information about the dataset
print(iris.info())
#plot with No. of data points on x axis
plt.plot(iris['species'])
plt.xlabel("No. of data points")
plt.show()

#histogram plot for iris['species']
plt.hist(iris["species"],color="green")
plt.show()

#To observe more about the data we can see a basic description of the data
iris.describe()

#scatter plot :sepal_length vs sepal_width.
sb.set_style('whitegrid')
sb.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
plt.plot()

#pairplot

sb.set_style('whitegrid')
sb.pairplot(iris,hue='species',size=3)
plt.show()

#Probability Density Function,Cumulative Distribution Function and HISTOGRAM
iris_setosa=iris.loc[iris['species']=='setosa']
iris_versicolor=iris.loc[iris['species']=='versicolor']
iris_virginica = iris.loc[iris['species']=='virginica']

plt.plot(iris_setosa['petal_length'],np.zeros_like(iris_setosa['petal_length'],),'o',label='setosa')
plt.plot(iris_virginica['petal_length'],np.zeros_like(iris_virginica['petal_length'],),'o',label='virginica')
plt.plot(iris_versicolor['petal_length'],np.zeros_like(iris_versicolor['petal_length'],),'o',label='versicolor')

plt.xlabel("petal_length")
plt.grid()
plt.legend()
plt.show()

#probability density function
sb.FacetGrid(iris,hue="species").map(sb.distplot,'petal_width').add_legend()
plt.show()
sb.FacetGrid(iris,hue="species").map(sb.distplot,'sepal_length').add_legend()
plt.show()
sb.FacetGrid(iris,hue="species").map(sb.distplot,'sepal_width').add_legend()
plt.show()
#petal_length>petal_width>>>sepal length>>>sepal width

#Cumulative Distribution Function

counts,bin_edges=np.histogram(iris_setosa['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf,label='pdf')
plt.plot(bin_edges[1:],pdf,label='cdf')

plt.legend()
plt.show()

```
